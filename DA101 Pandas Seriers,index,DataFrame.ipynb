```python
import pandas as pd #Importing pandas libarary

#Step to create your first data_frame : Two dimensional made of columns and rows : Attrbutes: Size Mutable,Heterogeneous,Labeled Axis
data = {
    'first_name':['Ram','Dorna','Willams'],
    'last_name':['Yadav','Gyawali','Windsor'],
    'email':['ram@ayodhya.com','dorna@aham.com','willams@buckingham.com']
}

df=pd.DataFrame(data)
df


```

```python
#changing the index
df.set_index('email',inplace=True)
```

```python
#usecase of creating and changing index is that we can easily locate 
# our data related to our index in this case we got a data related to Ram last_name which is Yadav syntax:  DataFrame.loc['index','column'] || optional ['column]

df.loc['ram@ayodhya.com','last_name']
```

```python
# indexes can be access through integer or numbers  and intergers are allowed  syntax DataFrame.iloc['0','1'] 0:index,1:Column is optional
df.iloc[0]
```

```python
# Resetting the index
df.reset_index(inplace=True)
df
```

Series and DataFrame

```python
# What is series in pandas : one dimensional array like that holds the data of any kind.
data=['Kathmandu','Butwal','Lipulekh']
s= pd.Series(data)
s
```

```python
#We can create our custom index 
data=['Kathmandu','Butwal','Lipulekh']
custom_index=['a','b','c']
s= pd.Series(data,index=custom_index)
s
```

```python
# series can be created by dict 
data={'NPL':'234560.00','IPL':'234235.00','GPL':'12.00'}
c_data=pd.Series(data)
c_data
```
