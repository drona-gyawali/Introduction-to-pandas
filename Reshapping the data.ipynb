### Titanic Dataset

1. **Reshape**: Convert the `Survived` and `Pclass` columns into a wide format table where each unique `Pclass` is a column, and the rows represent the `Survived` values.
2. **Melt**: Melt the columns `Pclass`, `Sex`, and `Survived` into a long format. Keep `PassengerId` as the identifier variable.
3. **Pivot**: Create a pivot table to find the average `Age` and `Fare` for each `Pclass` and `Sex` combination.
4. **Stack and Unstack**: Group by `Pclass` and `Survived`, calculate the mean of `Age` and `Fare`, then stack the result and unstack it to get back to the original table.
5. **Groupby**: Group the data by `Pclass` and calculate the average `Fare` for each class.

```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv')
df.head()
```

1. **Reshape**: Convert the `Survived` and `Pclass` columns into a wide format table where each unique `Pclass` is a column, and the rows represent the `Survived` values.


```python
reshape=df.pivot_table(columns='Pclass',index='Survived',aggfunc='size',fill_value=0) # 'size' represent the count so it makes sense and fill_value=0 handell the nan values
reshape
```

2. **Melt**: Melt the columns `Pclass`, `Sex`, and `Survived` into a long format. Keep `PassengerId` as the identifier variable.


```python
melting=pd.melt(df,
    id_vars=['PassengerId'],
    value_vars=['Sex','Pclass','Survived'],
    var_name='Attribute',
    value_name='Value'
)
melting
```

3. **Pivot**: Create a pivot table to find the average `Age` and `Fare` for each `Pclass` and `Sex` combination.


```python
avg_datas=df.pivot_table(index=['Sex','Pclass'],aggfunc={'Age':'mean',"Fare":'mean'})
avg_datas
```

4. **Stack and Unstack**: Group by `Pclass` and `Survived`, calculate the mean of `Age` and `Fare`, then stack the result and unstack it to get back to the original table.


```python
stacking=df.groupby(['Pclass','Survived']).agg({'Age':'mean','Fare':'mean'})
did_stack=stacking.stack(level=0)
stacked=pd.DataFrame(did_stack)
stacked
```

```python
did_stack.unstack(level=1)
```

5. **Groupby**: Group the data by `Pclass` and calculate the average `Fare` for each class.

```python
groupping=df.groupby('Pclass')['Fare'].mean()
pd.DataFrame(groupping)
```
