```python
import pandas as pd
dataset = {
    'Name': ['Alice', 'Bob', 'Charlie', None, 'Eva', 'Frank', 'Grace', 'Hannah', 'Ian', 'Jake', 'Kara', 'Louis', None, 'Mia', 'Nina',
             'Oscar', 'Paul', 'Quinn', 'Rachel', 'Steve', 'Tom', None, 'Uma', 'Victor', 'Wendy', 'Xander', 'Yara', 'Zane', None, 'Amy'],

    'Age': [25, None, 30, 22, 27, 'thirty', 29, 35, None, 40, 45, 'N/A', 33, 20, None, 31, 'unknown', 28, 25, 26, None, 32, 21, 23, 34, None, 'unknown', 27, 19, None],

    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', None, 'Phoenix', 'Philadelphia', None, 'San Antonio', 'San Diego',
             'Dallas', None, 'San Francisco', 'Austin', 'Columbus', 'Charlotte', 'Indianapolis', 'Seattle', None, 'Denver',
             'Washington', 'Boston', 'El Paso', 'Detroit', None, 'Nashville', 'Memphis', 'Portland', 'Baltimore', None],

    'Score': [85.5, None, 75.0, 90.5, 60.0, 'N/A', 88.0, 70.0, None, 95.0, None, 50.0, 82.0, None, 65.0, 78.0, None, 'missing', 60.0, 73.0, 85.5, None, 55.0, 91.0, 80.0, None, 'missing', 87.0, 75.0, None],

    'Employment Status': ['Employed', 'Unemployed', None, 'Employed', 'Student', 'Unemployed', 'Retired', 'Employed', 'Employed', None,
                          'Unemployed', 'Employed', 'Retired', 'Student', None, 'Employed', 'Employed', 'Student', 'Unemployed', 'Employed',
                          None, 'Unemployed', 'Student', 'Employed', 'Retired', None, 'Employed', 'Student', 'Retired', 'Employed']
}

df=pd.DataFrame(dataset)
```

```python
df.head()
```

```python
#Removing the row where index is None- Because 'Name' is an individual Variable of the datasets
df=df.dropna(subset=['Name'])
df.head()
```

```python
#Same operation with age but we will do some numerical thing with age column
# Lets define a fx of replacing value because there may be lots of replacement to do

def replace_value(df,col,old,new):
    df[col]=df[col].replace(old,new)

#Use cases of fx

replace_value(df,'Age','thirty','30') # Successfully replced the value

#We have to make sure that the age column is in int dtype

df['Age']=pd.to_numeric(df['Age'],errors='coerce')

#counting the nan value
nan_count=df['Age'].isna().sum()

#Rounded mean
if nan_count>0:
    mean_age=round(df['Age'].mean())

#Filling the nan values
df['Age']=df['Age'].fillna(mean_age)
df.head()
```

```python
# As col=City have some missing val so we are using mode for a frqu city to adjust the data 

mode_city=df['City'].mode()[0]
df['City']=df['City'].fillna(mode_city)
df.head()
```

```python
#As score play a crucial role in datasets.So we are going to change the none/n/a to '0'
df['Score']=df['Score'].fillna(0.0)

#applying the replace fx that we made earlier
replace_value(df,'Score','N/A','0.0')
replace_value(df,'Score','missing','0.0')
df.head()
```

```python
#In the col='Employment Status':  category data fulfill the requirement of the question . But there is null value so we have to replace it with 'Not defined' for clarity
df.fillna({'Employment Status':'Not defined'},inplace=True)
df

```
