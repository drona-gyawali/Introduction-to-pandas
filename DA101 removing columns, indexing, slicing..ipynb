```python
import pandas as pd
people={
    'first_name':['Dornaraj','Alex','Mavuer'],
    'last_name':['Gyawali','Zander','babbge'],
    'email':['drona@aham.com','Zander@alex.com','wollen@morgan.com']
}
df=pd.DataFrame(people)
df
```

```python
#Concating two words
df['first_name'].str.cat(df['last_name'], sep=" ") 
```

```python
#Adding new column
df['full_name']=df['first_name'].str.cat(df['last_name'], sep=" ") 
df
```

```python
#Deleting the column 
df.drop(columns=['first_name','last_name'], inplace=True) 
df
```

```python
#Spliting and inserting  the column
df[['first','last']]=df['full_name'].str.split(' ',expand=True)
df
```

```python
#adding the column only for first
new_row=pd.DataFrame({'first':['Batsman']})
df=pd.concat([df,new_row],ignore_index=True) #ignore_index is nessecary to excuete without error
df
```

```python
#Creating a dataframe and concating it to df
info={
    'email':['a@gmail.com','b@gmail.com','c@gmail.com'],
    'full_name':['Stark man','Batsman1','Robert 2.0']
}
df1=pd.DataFrame(info)
df=pd.concat([df1,df],ignore_index=True)  
df
```

```python
#Sort the values in desc form
df.sort_values(by ='full_name',ascending=True,inplace=True)
df
```

```python
#Deleting the specific row
df.drop(index=6) #Batsman has been removed
```

```python
#droping the row
df.drop(index=df[df['last']=='Zander'].index)
```
