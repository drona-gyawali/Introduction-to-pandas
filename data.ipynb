### Practical Implementation of Statstics 

## Measures of central tendency
1. Mean
2. Median
3. Mode


```python
import numpy as np
import statistics as s 
import math
import seaborn as sns

raw_data=[1,2,34,4,35,67,53,34,2,3,42,34,2,100]
print('Mean:' ,np.mean(raw_data))
print('Median:' ,np.median(raw_data))
print('Mode:' ,s.mode(raw_data))

```

## Outlier Detection

```python
sns.boxplot(raw_data)
```

### Five number summary

```python
q1,q3=np.percentile(raw_data,[25,75])

iqr=q3-q1
lower_fence=q1-1.5*iqr
higher_fence=q3+1.5*iqr

print(lower_fence)
print(higher_fence)
```

### Measure of Dispersion

1. Variance
2. Standard Deviation


```python
print('Sample Variance:',s.variance(raw_data)) # Sample variance
print('Population Variance:',np.var(raw_data)) # population variance
print('Standard Dev. BY Sample variance.:',s.stdev(raw_data))
print('Std. dev. by Population Variance',math.sqrt(s.pvariance(raw_data)))
```

### First std. dev to the left and right

```python
std= 29.883749121193972
mean=np.mean(raw_data)
print('std dev first right:', std+mean)
print('std. dev first left:',std-mean)
```

### Histogram and Pdf


```python
sns.histplot(raw_data,kde=True)
```

### Proving that the data follows the Gaussian Distribution
- If the probplot is staright to upward then thats Normal Distribution

```python
### q-q plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample,kde=True)
    plt.subplot(1,2,2)
    stat.probplot((sample),dist='norm',plot=pylab)
    plt.show()
```

```python
s=np.random.normal(0.5,0.2,1000)
plot_data(s)
```

```python
plot_data(iris_data['Sepal Width'])
```
