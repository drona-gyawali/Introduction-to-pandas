Group by and aggerate funtions

```python
ds= pd.read_csv('Flavors.csv')
ds
```

```python
group_by_frame= ds.groupby('Base Flavor').mean(['Flavor Rating','Texture Rating','Total_Rating'])
group_by_frame
```

```python
count_all=ds.groupby('Base Flavor').count()
count_all
```

```python
minimum=ds.groupby('Base Flavor').min()
print(minimum)
maximum=ds.groupby('Base Flavor').max()
print(maximum)
```

```python
total = ds.groupby('Base Flavor').sum(['Flavor Rating','Texture Rating','Total Rating'])
total
```

```python
all_in_one= ds.groupby('Base Flavor').agg({'Flavor Rating':['mean','sum','min','max']})
all_in_one
```

```python
ds.groupby(['Base Flavor', 'Liked']).mean(['Flavor Rating','Texture Rating','Total_Rating'])

```

```python
ds.groupby(['Base Flavor', 'Liked']).agg({'Flavor Rating': ['mean','sum','max','count']})
```

```python
ds.groupby(['Base Flavor', 'Liked']).describe()
```

Merge, Join & Concat

```python
df1= pd.read_csv('LOTR.csv')
df1
```

```python
df2= pd.read_csv('LOTR 2.csv')
df2
```

Inner join

```python
df1.merge(df2, how='inner', on = ['FellowshipID','FirstName'])
```

Outer Join

```python
df1.merge(df2, how = 'outer')
```

left join

```python
df1.merge(df2,how ='left')  # Trying to join table from left (left table) left to right
```

right join

```python
df1.merge(df2, how ='right')  # Trying to join table from right  (right table) right to left
```

Cross join

```python
df1.merge(df2,how = 'cross')
```

join

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix='_left',rsuffix='_right',how ='outer')
df4
```

Concatinate

```python
pd.concat([df1,df2],join='outer',axis=1)
```

```python
beta=pd.read_excel('Customer Call List.xlsx')
beta
```

```python
#deleting the duplicates
beta = beta.drop_duplicates()
beta
```

```python
#removing the columns
beta=beta.drop(columns='Not_Useful_Column')
beta
```

```python
# triming special characters
beta['Phone_Number']= beta['Phone_Number'].str.replace(r'[^\d]','', regex=True)
beta
```

```python
# step 1. converting the data into str
beta['Phone_Number']= beta['Phone_Number'].apply(lambda x : ''.join(filter(str.isdigit,str(x))) if pd.notna(x) else x)
#step 2 - placing the - in the numbers between to format xxx-xxx-xxxx 
beta['Phone_Number']=beta['Phone_Number'].apply(lambda x : f'{x[0:3]}-{x[3:6]}-{x[6:10]}' if pd.notna(x) and len(x)==10 else x)
beta
```

```python
#Spilting the column
spilt_column=beta['Address'].str.split(',',n=2,expand=True)
beta[['Street Address','State','zip_code(opt)']]=spilt_column
beta
```

```python
#droping the column
beta=beta.drop(columns='Address')
beta
```

```python
# Standarlzing the value
beta['Paying Customer']=beta['Paying Customer'].replace({'Y':'Yes','N':'No'})

beta['Do_Not_Contact']=beta['Do_Not_Contact'].replace({'Y':'Yes','N':'No'})
beta
```

```python
#Removing unwanted special characters in last name using regex
pattern = r'[^a-zA-Z0-9]'
beta['Last_Name']=beta['Last_Name'].str.replace(pattern,'',regex=True)
beta
```

```python
beta=beta.fillna('')
beta
```

```python
# Identifying the indexes to drop using list comprehension
index_drop= [x for x in beta.index if beta.loc[x,'Do_Not_Contact']=='Yes']

# Drop the index
beta.drop(index_drop,inplace=True)
beta
```

```python
beta['Do_Not_Contact'] =beta['Do_Not_Contact'].replace(['', ' ', None], 'No')
beta
```

```python
# droping the customer without phone numbers
index_drop= [x for x in beta.index if beta.loc[x,'Phone_Number']=='']

# removing the data
beta.drop(index_drop, inplace=True)
beta
```
