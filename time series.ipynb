## Two types of Datetime objects in python
- **Naive (No timezone awarness)**
- **Time Zone aware datetime**

```python
beta.index
```

```python
# Convert index to datetime if necessary
if not isinstance(beta.index, pd.DatetimeIndex):
    beta.index = pd.to_datetime(beta.index)

# Localize timezone if it is naive
if beta.index.tz is None:
    beta = beta.tz_localize(tz='US/Eastern')
```

```python
beta.index
```

```python
beta=beta.tz_convert(tz='Asia/Kathmandu')
beta
```

```python
rng=pd.date_range(start='2017-08-17 18:45:00',periods=10,freq='30min')
s=pd.Series(range(10),index=rng)
s
```

```python
Nepal=s.tz_localize(tz='Asia/Katmandu')
us=s.tz_localize(tz='US/Eastern')
print(Nepal-us)
```
