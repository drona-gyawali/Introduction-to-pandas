````python
import pandas as pd

# Create Orders DataFrame
orders_data = {
    'order_id': [1, 2, 3, 4, 5],
    'customer_id': [101, 102, 101, 103, 104],
    'product_id': ['P01', 'P02', 'P03', 'P01', 'P04'],
    'order_date': ['2024-10-01', '2024-10-03', '2024-10-04', '2024-10-05', '2024-10-06'],
    'quantity': [2, 1, 3, 1, 5],
    'order_amount': [400, 150, 900, 200, 1000]
}
orders = pd.DataFrame(orders_data)

# Create Customers DataFrame
customers_data = {
    'customer_id': [101, 102, 103, 104, 105],
    'customer_name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'location': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Miami']
}
customers = pd.DataFrame(customers_data)

# Create Products DataFrame
products_data = {
    'product_id': ['P01', 'P02', 'P03', 'P04', 'P05'],
    'product_name': ['Phone', 'Laptop', 'Tablet', 'Headphones', 'Camera'],
    'price': [200, 150, 300, 200, 500]
}
products = pd.DataFrame(products_data)

# Display datasets
print("Orders:\n", orders)
print("\nCustomers:\n", customers)
print("\nProducts:\n", products)
````

1. **Task 1: Merge (Inner Join)**  
  Merge the **Orders** and **Customers** datasets based on `customer_id` to create a DataFrame that shows each order along with the customer details (name and location).

````python
import pandas as pd
merge_table_inner= pd.merge(orders,customers,on='customer_id',how='inner' )
merge_table_inner.head()
````

2. **Task 2: Merge (Left Join)**  
   Perform a **left join** between the **Orders** and **Products** datasets using `product_id`. The resulting DataFrame should show each order along with the product details (name and price).


````python
merge_table_left= pd.merge(orders,products,on='product_id',how='left')
merge_table_left.head()
````

3. **Task 3: Concatenation**  
   Add the following additional orders to the original **Orders** DataFrame by concatenating them:

````python
import pandas as pd

concat_data = {
    'order_id': [6, 7],
    'customer_id': [105, 102],
    'product_id': ['P05', 'P03'],
    'order_date': ['2024-10-07', '2024-10-08'],
    'quantity': [1, 2],
    'order_amount': [500, 600]
}

concat_data = pd.DataFrame(concat_data)

all_orders=pd.concat([orders,concat_data],ignore_index=True)
all_orders
````

Task 5: Merge with Missing Data (Right Join)
Perform a right join between the Orders and Products datasets using product_id. This should show all products, including those that were not ordered.

````python
missin_datas=pd.merge(orders,products,how='right',on='product_id')
missin_datas
````
